```polyglot-notebook
ALTER TABLE SQLBooks
ADD embeddings_2 VECTOR(1024)
```

```polyglot-notebook
SET NOCOUNT ON
DECLARE @BatchSize INT = 150

WHILE 1 = 1
BEGIN
    ;WITH CTE AS (
        SELECT TOP (@BatchSize) *
        FROM SQLBooks
        WHERE embeddings_2 IS NULL
        ORDER BY id 
    )
    UPDATE CTE
    SET embeddings_2 = AI_GENERATE_EMBEDDINGS(title USE MODEL ollama2)

    IF @@ROWCOUNT < @Batchsize
        BREAK
END
SET NOCOUNT OFF
```
